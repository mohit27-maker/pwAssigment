1. Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

 >Artificial Intelligence (AI)
 - AI is the broad field of creating machines that can perform tasks requiring human-like intelligence.
 - Make systems that can think, reason, learn, and make decisions like humans.
 - Very wide – includes rule-based systems, expert systems, robotics, NLP, ML, DL, etc.
 - Example: A chatbot that answers customer queries, chess-playing computers, or self-driving cars.

 >Machine Learning (ML)
 -  ML is a subset of AI that enables machines to learn from data and improve performance without explicit programming.
 -  Build models that can find patterns in data and make predictions or decisions.
 -  Part of AI; includes supervised, unsupervised, and reinforcement learning.
 - Example: Spam email detection, predicting stock prices, movie recommendations.

 >Deep Learning (DL)
 - DL is a subset of ML that uses artificial neural networks with many layers (deep neural networks).
 -  Handle large, complex datasets like images, videos, and natural language, automatically extracting features.
 -  Advanced part of ML; needs huge amounts of data and computing power.
 - Example: Face recognition, voice assistants (Siri, Alexa), autonomous driving vision systems.

 >Data Science (DS)
 -  DS is the interdisciplinary field that combines statistics, data analysis, ML, and domain knowledge to extract insights from data.
 - Turn raw data into useful information for decision-making.
 -  Broader than AI/ML/DL – includes data cleaning, visualization, reporting, and building models.
 - Example: Analyzing sales data to recommend strategies, customer segmentation, fraud detection dashboards.

2. What are the types of machine learning? Describe each with one
real-world example.

> Supervised Learning
 - The model is trained on a labeled dataset (input → output). It learns the mapping from input features to the correct output.
 -  Predict outcomes for new, unseen data.
- Two types:
 a- Regression → Predict continuous values.
 b- Classification → Predict categories.

Example:
- Predicting house prices based on size, location, and number of rooms.
- Detecting spam vs. non-spam emails.

> Unsupervised Learning
 - The model is trained on an unlabeled dataset (no predefined output). It tries to find patterns, clusters, or structure in the data.
 - Discover hidden structures or groupings in data.

Two common tasks:

- Clustering → Grouping similar data points.
- Dimensionality Reduction → Simplifying data while keeping important info.

Example:

- Customer segmentation in marketing (grouping buyers into clusters like “budget shoppers” vs. “luxury shoppers”).
- Compressing high-dimensional image data to fewer features for faster processing.

> Reinforcement Learning (RL)
 - The model (called an agent) learns by interacting with an environment and receiving rewards or penalties for its actions.
 -  Maximize cumulative reward by choosing the best sequence of actions.

Example:

- Self-driving cars: The agent learns to drive safely by getting positive rewards for staying in lane and negative rewards for collisions.
- Game AI: Training an AI to play chess or Go (AlphaGo by DeepMind).

3. Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.

 > Overfitting
- When a model learns the training data too well, including its noise and outliers, but fails to generalize to new unseen data.
- Model is too complex (too many parameters, deep trees, too many layers).

Symptoms:
- Very high accuracy on training data.
 - Poor performance on test/validation data.

 - Example: A student memorizes past exam answers word-for-word but struggles with slightly different questions in the real exam.

 > Underfitting

-  When a model is too simple to capture the patterns in the data, leading to poor performance on both training and test data.

- Model has too few parameters, wrong algorithm, or not enough training.

Symptoms:
 - Low accuracy on both training and test sets.

- Example: A student only skims through one chapter before the exam and cannot answer even the basic questions properly.

 > Bias-Variance Tradeoff

 - Bias = Error due to overly simplistic assumptions in the model (leads to underfitting).
 - Variance = Error due to too much sensitivity to training data (leads to overfitting).

Tradeoff:
 - If bias is high → the model is too simple (can’t capture patterns).
 - If variance is high → the model is too complex (captures noise as patterns).

- The goal is to find the sweet spot between bias and variance for good generalization.

Analogy:
 - High bias, low variance → All arrows are tightly grouped but far from the target center.
 - Low bias, high variance → Arrows are spread out all over, sometimes hitting the target, sometimes missing.

- Low bias, low variance → Arrows are tightly grouped and close to the bullseye (ideal).

4. What are outliers in a dataset, and list three common techniques for
handling them.
 > Outliers: Definition
 - An outlier is a data point that is significantly different from other observations in the dataset.
 - They may occur due to:
   
   a) Data entry errors (typos, sensor failures).
   
   b) Natural rare events (fraudulent transaction, extreme weather).
 - Variability in measurement.

- Impact:

 - Can skew averages and variances, leading to misleading analysis.

 - Can negatively affect model performance (especially regression models).

Example:
If most people's monthly income is between ₹20,000–₹80,000, but one record shows ₹9,00,000, that’s an outlier.

> Three Common Techniques for Handling Outliers

1. Removal (Deleting Outliers)
 - If outliers are due to data entry errors or are irrelevant to the analysis, they can be removed.
 - Method: Use statistical rules like
 - Z-Score: Data points beyond ±3 standard deviations.
 - IQR (Interquartile Range)

Example: Removing impossible ages like 150 years in a “customer age” column.

> Transformation (Reducing Outlier Impact)
 - Apply mathematical transformations to reduce the influence of extreme values.
 - Techniques:

  a) Log transformation

  b) Square root transformation

  c) Box-Cox transformation

Example: Skewed income data can be log-transformed to reduce the gap between low and high values.

> Imputation or Capping (Winsorization)

- Replace outliers with a more reasonable value instead of deleting.
- Methods:
- Replace with mean/median.
- Cap values at certain percentiles (e.g., replace all values above the 95th percentile with the 95th percentile value).

Example: In housing price data, extreme luxury house prices could be capped at the 99th percentile.

5. : Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.

**Process of Handling Missing Values**

1. Identify Missing Data
- Check how many and where missing values exist.
- Tools: .isnull(), .sum() in Pandas.

2. Analyze the Pattern of Missingness
- MCAR (Missing Completely at Random): No relationship with any other variable.
- MAR (Missing at Random): Related to other observed variables.
- MNAR (Missing Not at Random): Missingness depends on the missing value itself.

3. Decide a Strategy
- Remove → If too many missing values (e.g., >50%).
- Impute (Fill in) → Replace with a suitable value.
- Leave as Missing → Sometimes useful for tree-based models.

4. Apply Appropriate Technique
- Based on data type (numerical / categorical) and situation.

**Imputation Techniques**
> For Numerical Data → Mean/Median Imputation
- Replace missing values with the mean (for normal distribution) or median (for skewed distribution).
- Example: If “Age” has missing values, fill them with the median age of the dataset.

> For Categorical Data → Mode Imputation
- Replace missing values with the most frequent category (mode).
- Example: If “Gender” column has missing values, fill them with the most common gender in the dataset.

```python
#: Write a Python program that:
# ● Creates a synthetic imbalanced dataset with make_classification() from
# sklearn.datasets.
# ● Prints the class distribution
from sklearn.datasets import make_classification
from collections import Counter
import pandas as pd

X, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=5,
    n_redundant=2,
    n_classes=2,
    weights=[0.9, 0.1],
    random_state=42
)

class_distribution = Counter(y)
print("Class Distribution:", class_distribution)

df_distribution = pd.DataFrame.from_dict(class_distribution, orient='index', columns=['Count'])
df_distribution['Percentage'] = df_distribution['Count'] / df_distribution['Count'].sum() * 100
print("\nClass Distribution (with percentage):\n", df_distribution)
```

```python
# Write a Python script to:
# ● Generate 1000 samples from a normal distribution.
# ● Introduce 50 random missing values.
# ● Fill missing values with the column mean.
# ● Plot a histogram before and after imputation.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

data_missing = data.copy()
missing_indices = np.random.choice(1000, 50, replace=False)
data_missing[missing_indices] = np.nan

df = pd.DataFrame(data_missing, columns=['Value'])
mean_value = df['Value'].mean()
df['Imputed'] = df['Value'].fillna(mean_value)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['Value'].dropna(), bins=30, edgecolor='black')
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(df['Imputed'], bins=30, edgecolor='black')
plt.title("After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

10. : You are working as a data scientist for a retail company. You receive a customer
transaction dataset that contains:
● Missing ages,
● Outliers in transaction amount,
● A highly imbalanced target (fraud vs. non-fraud),
● Categorical variables like payment method.
Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.

1. Handle Missing Data (e.g., Missing Ages)

- Identify Missingness
- Use .isnull().sum() to see how many missing values exist.
 > Imputation Strategy:
 - For numerical features (like Age):
- If missing is small → Fill with mean/median (median preferred if skewed).
- If missing is large → Use predictive imputation (e.g., train a regression/KNN model using other features like location, income).
- For categorical features (if any missing): Fill with mode or add a new category "Unknown".


2. Handle Outliers (Transaction Amount)

- Detect Outliers:
- Use IQR method: values beyond Q1 - 1.5*IQR or Q3 + 1.5*IQR.
- Or use Z-score (>3 standard deviations).

> Treatment Options:
- If outlier is a data entry error (e.g., transaction = ₹1,000,000,000) → Remove it.
- If outlier is genuine but extreme → Cap (Winsorization) at the 1st/99th percentile.
- Alternatively → Apply log transformation to reduce skewness.

3. Handle Class Imbalance (Fraud vs. Non-Fraud)


- Problem: Fraud cases are usually very rare (e.g., 1% fraud, 99% non-fraud).

> Solutions:

 Resampling:

- Oversample minority (fraud) → e.g., SMOTE (Synthetic Minority Over-sampling Technique).
- Undersample majority (non-fraud).

 Algorithm-level:

- Use class weights in models (e.g., class_weight='balanced' in logistic regression, random forest, XGBoost).

Evaluation Metrics:

- Avoid accuracy (misleading). Use Precision, Recall, F1-score, ROC-AUC, PR-AUC.


4. Handle Categorical Variables (e.g., Payment Method)

Encoding Strategy:

- If low-cardinality categorical variable (few categories, like "Credit Card", "UPI", "Cash") → Use One-Hot Encoding.

- If high-cardinality variable (hundreds of categories, e.g., product ID) → Use Target Encoding / Frequency Encoding / Embedding.


5. Scaling Numerical Features (Optional but Recommended)
- Some models (Logistic Regression, Neural Networks, KNN) need scaled data.
- Use StandardScaler (z-score normalization) or MinMaxScaler (0–1 scaling).
- Tree-based models (Random Forest, XGBoost) don’t require scaling.

6. Train-Test Split

- Ensure stratified split to maintain fraud/non-fraud ratio in both train & test sets.

**Final Workflow Summary**
- Missing Data → Impute ages (median for numerical, mode/“Unknown” for categorical).
- Outliers → Detect with IQR/Z-score; cap or transform.
- Imbalance → Use SMOTE/undersampling + class-weighting; evaluate with ROC-AUC, Precision-Recall.
- Encoding → One-hot encode payment method (or target encoding for high-cardinality).
- Scaling → Apply StandardScaler/MinMaxScaler (if model needs it).
- Split Data → Stratified train-test split.
