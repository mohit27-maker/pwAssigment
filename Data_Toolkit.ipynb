```python
# How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
arr = np.array([[1,2,3],
                 [3,4,5],
                 [6,7,8]])
row_sum = arr.sum(axis=1)   #arr.sum(axis=0) → Column-wise sum
                            #arr.sum(axis=1) → Row-wise sum
print(row_sum)
```

```python
# Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
data = {'name': ['aman','rahul','vikas'],
        'age' : [27,25,35],
        'salary' : [2500,3600,4900]}
df = pd.DataFrame(data)

mean_age = df['age'].mean()
print(mean_age)
```

```python
# A Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
x = [6,7,9,7,6,5,8]
y = [96,88,101,79,103,81,83]

plt.scatter(x,y,color = 'blue',marker = 'o')

plt.xlabel('x value')
plt.ylabel('y value')
plt.title('simple scatter plot')

plt.show()
```

```python
#  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'sge': [22,29,36,40,45],
        'salary':[2000,2500,3000,4000,4500],
        'experience':[2,4,6,8,9]}

df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix,annot= True ,cmap = 'coolwarm',linewidths = 0.5 ) # Blue for negative correlation,
                                                                          # Red for positive correlation,
                                                                          # White near zero correlation

plt.title('correlation matrix heatmap')
plt.show()
```

```python
#  Generate a bar plot using Plotly.
import pandas as pd
import plotly.express as px

data = {'fruits' : ['apple','banana','guava','anar'],
        'quantity' : [10,11,28,5]}
df = pd.DataFrame(data)

fig = px.bar(df,x='fruits',y='quantity',title='fruit quantity bar plot',color='fruits')
fig.show()
```

```python
# Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 3, 4, 5, 6]

plt.plot(x, y1, label = 'line1',marker = 'd')
plt.plot(x, y2, label = 'line2',marker = '*')
plt.plot(x, y3,label = 'line3',marker = "^")

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("line plot with multiple lines")
plt.legend()

plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

data = {'name':['alice','max','joe','mike'],
        'marks' :[89,75,95,66]}
df = pd.DataFrame(data)

filtered_df = df[df['marks']>80]

print("raw data :\n",df)
print("filtered data :\n",filtered_df)
```

```python
# Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

data = [12,15,20,25,13,19,16,23,22,19,15,23]

sns.histplot(data,bins=8,kde=True,color='blue')

plt.xlabel('values')
plt.ylabel('frequence')
plt.title('histogram of data distribution')

plt.show()
```
