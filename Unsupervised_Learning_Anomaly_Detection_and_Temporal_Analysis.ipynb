Question 2: Name and briefly describe three common dimensionality reduction
techniques..

- Three Common Dimensionality Reduction Techniques

1. PCA (Principal Component Analysis):
- Reduces features by creating new uncorrelated components that capture maximum variance. Widely used for speeding up computation and removing redundancy.

2. t-SNE (t-Distributed Stochastic Neighbor Embedding):
- Nonlinear method for visualizing high-dimensional data in 2D or 3D, preserving local structure and clustering similar points.

3. LDA (Linear Discriminant Analysis):
- Supervised method that reduces dimensions while maximizing class separability, often used in classification tasks.

Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms

- Clustering is an unsupervised learning technique that groups similar data points together based on patterns or similarities in the data, without using labeled outcomes. The goal is to discover natural groupings or structures within the dataset.

> Three Popular Clustering Algorithms

1. K-Means Clustering:
- Partitions data into K clusters by minimizing the distance between points and their cluster centroids.

2. Hierarchical Clustering:
- Builds a tree-like structure (dendrogram) of nested clusters, either by merging (agglomerative) or splitting (divisive) clusters.

3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise):
- Groups points that are densely packed together and identifies outliers as noise, useful for clusters of arbitrary shapes.

Question 4: Explain the concept of anomaly detection and its significance.

4. Security: Flags suspicious activities in cybersecurity.

Anomaly detection is crucial because detecting rare but significant events can prevent losses, improve reliability, and enhance decision-making.

Question 5: List and briefly describe three types of anomaly detection techniques.

- Three Types of Anomaly Detection Techniques

1. Statistical Methods:
These methods assume that the data follows a known statistical distribution, such as Gaussian (normal) distribution. Anomalies are detected as data points that significantly deviate from this expected distribution. Common techniques include Z-score, Grubbsâ€™ test, and probability-based approaches. Statistical methods are simple and effective for well-behaved datasets but may fail if the data distribution is unknown or complex.

2. Machine Learning Methods:
Machine learning approaches use algorithms to learn patterns of normal behavior from historical data. Once the model is trained, it can identify data points that differ from these learned patterns as anomalies. Techniques include supervised learning (with labeled normal and anomalous data), unsupervised learning (e.g., One-Class SVM, Isolation Forest), and semi-supervised learning. These methods can handle complex and high-dimensional datasets.

3. Proximity-Based Methods:
These methods detect anomalies based on the distance or density of data points. Points that are far from others (distance-based) or in low-density regions (density-based) are considered anomalies. Examples include k-Nearest Neighbors (k-NN) and DBSCAN. Proximity-based methods are effective when anomalies are far from clusters but may struggle with varying densities in the data.

Question 6: What is time series analysis? Mention two key components of time series data.
- Time series analysis is the process of analyzing data points collected or recorded at successive time intervals to identify patterns, trends, and seasonal variations. It is used to forecast future values, understand past behavior, and make data-driven decisions in areas like finance, weather, stock markets, and sales.

> Two Key Components of Time Series Data

1. Trend:
The long-term movement or general direction in the data over time (e.g., increasing sales over several years).

2. Seasonality:
Regular, repeating patterns within a specific time period (e.g., higher ice cream sales every summer).

Other components sometimes considered include cyclical patterns and random noise, but trend and seasonality are the most fundamental.


Question 7: Describe the difference between seasonality and cyclic behavior in time series.

> Seasonality
- Regular, repeating patterns in a fixed period (e.g., daily, monthly, yearly).
- Predictable and short-term fluctuations.
- Caused by calendar events, holidays, or weather.
- Example: Ice cream sales increase every summer.

>Cyclic Behavior
- Fluctuations over a longer, variable period.
- Less predictable and not tied to a fixed calendar.
- Caused by economic or business cycles.
- Example: Economic growth and recession cycles over several years.


Question 8: Write Python code to perform K-means clustering on a sample dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.7, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', s=50, alpha=0.6)
plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.

- Inheritance is a fundamental concept in Object-Oriented Programming (OOP) where a class (called child/subclass) can inherit properties and methods from another class (called parent/superclass). This allows code reuse and hierarchical relationships between classes.

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound.")

# Child class inheriting from Animal
class Dog(Animal):
    def speak(self):
        print(f"{self.name} barks.")

# Create objects
a = Animal("Generic Animal")
d = Dog("Buddy")

# Call methods
a.speak()
d.speak()
```

Question 10: How can time series analysis be used for anomaly detection?

- Time series analysis can help detect anomalies by identifying data points that deviate from expected patterns over time. Since time series data has components like trend, seasonality, and noise, anomalies are points that do not follow these normal patterns.

>Approaches:
- Statistical Methods: Compare observed values with expected values based on historical trends or moving averages. Points outside confidence intervals are anomalies.
- Machine Learning Models: Models like LSTM, ARIMA, or Isolation Forest can learn normal time series behavior and flag unusual deviations.
- Seasonal Decomposition: Break the series into trend, seasonality, and residuals; large residuals can indicate anomalies.

>Applications:
- Detecting unusual spikes in website traffic.
- Identifying sudden equipment failures in IoT sensor data.
- Spotting fraudulent transactions over time.
