7. What is the significance of confidence intervals in statistical analysis?

 = Confidence Interval (CI)

A confidence interval gives a range of values that is likely to contain the true population parameter (like mean or proportion) with a certain level of confidence (e.g., 95%).

General Form : CI = Point Estimate Â± Margin of Error

> Significance in Statistical Analysis

1. Quantifies Uncertainty
 - Instead of giving just one estimate (sample mean), CI provides a range, showing the precision and reliability of the estimate.

2. Decision-Making Tool
 - Widely used in business, medicine, and research to determine if an effect is real.

3. Alternative to Hypothesis Testing
 - If a CI for a mean difference includes 0, it suggests no significant difference.

 - If a CI for a proportion includes the null value, the result is not statistically significant.

4. Connection to Probability
 - A 95% CI does not mean the parameter has a 95% chance of being inside the interval.
 - Instead, it means that if we took 100 random samples and built 100 confidence intervals, about 95 of them would contain the true parameter.

```python
# Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5.
# Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
std_dev = np.std(data)

print(f"Sample Mean: {mean:.2f}")
print(f"Sample Standard Deviation: {std_dev:.2f}")

plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title("Histogram of Normally Distributed Data (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean:.2f}")
plt.legend()
plt.show()
```

```python
# You are working as a data analyst for a retail company. The company has
# collected daily sales data for 2 years and wants you to identify the overall sales trend.
# daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
# 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
# â— Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
# â— Write the Python code to compute the mean sales and its confidence interval.

import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

data = np.array(daily_sales)


mean_sales = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)


std_error = std_dev / np.sqrt(n)

confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha/2)

margin_of_error = z_critical * std_error
ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

print(f"Sample Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
