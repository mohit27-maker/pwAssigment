Question 1: What is hypothesis testing in statistics?

 = Hypothesis testing in statistics is a method of making decisions or inferences about a population based on data collected from a sample.It helps us determine whether there is enough evidence to support a certain claim (hypothesis) about the population.

 * key steps in hypothesis testing :
 1. formulate hypothesis
 2. choose significance level(α)
 3. select the test and calculate the statistics
 4. find the key value or critical value
 5. make a decision


 Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

> Null Hypothesis (H₀):

- The default assumption in hypothesis testing.
- States that there is no effect, no difference, or no relationship in the population.
- It is what we try to test against.

>Alternative Hypothesis (H₁ or Ha):

- The opposite of H₀.
- States that there is an effect, a difference, or a relationship.
- This is usually the research hypothesis we want to prove.

key difference:

- The null hypothesis (H₀) assumes there is no effect or no difference; it represents the default or status quo.
- The alternative hypothesis (H₁/Ha) assumes there is an effect or a difference; it represents what the researcher wants to prove.


Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

 > Significance Level (α)
 - The significance level (denoted by α) is the threshold probability we set before conducting a hypothesis test.
 - It represents the maximum risk of rejecting the null hypothesis (H₀) when it is actually true (Type I error).
 - Common choices: 0.05 (5%), 0.01 (1%), 0.10 (10%).

> Role in Deciding the Outcome
- After calculating the p-value from the test:
1. If p ≤ α → Reject H₀ → evidence supports H₁.
2. If p > α → Fail to reject H₀ → insufficient evidence to support H₁.\


Question 4: What are Type I and Type II errors? Give examples of each.

 > Type I Error (False Positive)

- Happens when we reject the null hypothesis (H₀) even though it is true.
- It means we detect an effect that does not actually exist.
- Probability of making a Type I error = α (significance level).

Example:

- A court trial: H₀ = "The person is innocent".
- Type I error = Wrongly convicting an innocent person.
- In medical testing: H₀ = "Patient does not have a disease".
- Type I error = Test says the patient has the disease (but they don’t).


 > Type II Error (False Negative)

- Happens when we fail to reject the null hypothesis (H₀) even though it is false.
- It means we miss an effect that actually exists.
- Probability of making a Type II error = β.

Example:

- Court trial: H₀ = "The person is innocent".
- Type II error = Letting a guilty person go free.
- Medical testing: H₀ = "Patient does not have a disease".
- Type II error = Test says the patient is healthy (but they actually have the disease).


Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

 1. Z-Test
- A statistical test used when:
- The population variance (σ²) or standard deviation (σ) is known.
- The sample size is large (n > 30) → Central Limit Theorem makes data approximately normal.
- Based on the Z-distribution (standard normal, mean = 0, SD = 1).

2. T-Test
- A statistical test used when:
- the population variance (σ²) is unknown.
- The sample size is small (n ≤ 30).
- We estimate variance using the sample standard deviation (s).
- Based on the t-distribution, which has heavier tails than Z-distribution (more uncertainty with small samples).

Key differences:
- A Z-test is used when the population standard deviation is known and the sample size is large (n > 30), relying on the Z-distribution.
- A T-test is used when the population standard deviation is unknown and the sample size is small (n ≤ 30), relying on the t-distribution with heavier tails to handle more uncertainty.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
# 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
# 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
# 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu0 = 50
alpha = 0.05

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_stat = (sample_mean - mu0) / (sample_std / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", round(sample_mean, 3))
print("Sample Std Dev:", round(sample_std, 3))
print("Z Statistic:", round(z_stat, 3))
print("P-value:", round(p_value, 4))

if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
mu, sigma = 50, 5
n = 100
data = np.random.normal(mu, sigma, n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
alpha = 0.05
df = n - 1

t_crit = stats.t.ppf(1 - alpha/2, df)

margin_error = t_crit * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")

plt.hist(data, bins=15, color="skyblue", edgecolor="black", alpha=0.7)
plt.axvline(sample_mean, color="red", linestyle="dashed", linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color="green", linestyle="dashed", linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color="green", linestyle="dashed", linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = (data - mean) / std_dev
    return z_scores

data = np.array([10, 12, 13, 15, 18, 20, 21, 22, 25, 30])

z_scores = calculate_z_scores(data)

print("Original Data:", data)
print("Z-Scores:", np.round(z_scores, 2))

plt.hist(z_scores, bins=8, color="lightblue", edgecolor="black", alpha=0.7)
plt.axvline(0, color="red", linestyle="dashed", linewidth=2, label="Mean (Z=0)")
plt.title("Histogram of Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
